```python
for i in range(len(irisflower)):
    if irisflower['class'][i]=="iris_setosa":
       irisflower['class'][i]=0
    if irisflower['class'][i]=="iris_versicolor":
       irisflower['class'][i]=1
    if irisflower['class'][i]=="iris_virginica":
       irisflower['class'][i]=2
irisflower    
```

```python
import numpy as np
```

```python
 x=irisflower[['sepal_length','sepal_width','petal_length',' petal_width']]
```

```python
y=irisflower['class']
```

```python
x=x.astype(float)
x
```

```python
y=y.astype(float)
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,shuffle=True,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
lr.score(x_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
import sklearn
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print("report:\n",classification_report(y_test,y_pred))
```
